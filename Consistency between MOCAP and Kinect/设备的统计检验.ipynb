```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# 准备数据，假设有两个设备（设备A和设备B）在四个任务下的速度数据
mo_sp_2m=pd.read_csv("C:\\Users\\lyy\\Desktop\\Data\\Parameters\\MOCAP_2m\\spatio-temporal_parameters.csv")
ki_sp_2m=pd.read_csv("C:\\Users\\lyy\\Desktop\\Data\\Parameters\\Kinect_2m\\spatio-temporal_parameters.csv")
```

```python
#mocap提取first-go-task06
#第一次和第二次分开
mo_time = mo_sp_2m.groupby('times')
mo_time01 = mo_time.get_group('time01')
#直行和转弯阶段分开
mo_time01_process = mo_time01.groupby('process')
mo_time01_go  = mo_time01_process .get_group('go')
#单双任务分开
mo_time01_go_task = mo_time01_go.groupby('task')
mo_time01_go_normal = mo_time01_go_task.get_group('task_normal')


#kinect提取first-go-task06
#第一次和第二次分开
ki_time = ki_sp_2m.groupby('times')
ki_time01 = ki_time.get_group('time01')
#直行和转弯阶段分开
ki_time01_process = ki_time01.groupby('process')
ki_time01_go  = ki_time01_process .get_group('go')
#单双任务分开
ki_time01_go_task = ki_time01_go.groupby('task')
ki_time01_go_normal = ki_time01_go_task.get_group('task_normal')

```

```python

import numpy as np
merged = pd.concat([mo_time01_go_normal, ki_time01_go_normal], ignore_index=True)
merged
```

```python
import pingouin as pg
from scipy import stats
mo_step_normal = np.array(mo_time01_go_normal['step_length'])  # 第一列数据
ki_step_normal = np.array(ki_time01_go_normal['step_length']) # 第二列数据

step_st=stats.ttest_rel(mo_step_normal,ki_step_normal)

print(step_st)
# 计算ICC值

print(np.mean(mo_step_normal).round(3),"±",np.std(mo_step_normal).round(3))
print(np.mean(ki_step_normal).round(3),"±",np.std(ki_step_normal).round(3))
```

```python

mo_speed_normal = np.array(mo_time01_go_normal['stride_speed'])  # 第一列数据
ki_speed_normal = np.array(ki_time01_go_normal['stride_speed']) # 第二列数据

speed_st=stats.ttest_rel(mo_speed_normal,ki_speed_normal)

print(speed_st)
# 计算ICC值

print(np.mean(mo_speed_normal).round(3),"±",np.std(mo_speed_normal).round(3))
print(np.mean(ki_speed_normal).round(3),"±",np.std(ki_speed_normal).round(3))
```

```python
import numpy as np
import pingouin as pg
from scipy import stats
data1 = np.array(mo_time01_go_normal['step_length'])  # 第一列数据
data2 = np.array(ki_time01_go_normal['step_length']) # 第二列数据
data3 = np.array(mo_time01_go_normal['stride_speed'])  # 第一列数据
data4 = np.array(ki_time01_go_normal['stride_speed']) # 第二列数据
custom_font = 'Times New Roman'
plt.style.use('default')


plt.figure(figsize=(16, 6))

plt.subplot(1,2,1)
ax = pg.plot_blandaltman(data1, data2)
# 设置字体属性
font_properties = {'fontsize': 15,'fontname': custom_font}

# 修改图表中的文本元素
for text in ax.texts:
    text.set_fontsize(font_properties['fontsize'])
    text.set_fontname(font_properties['fontname'])

#ax = fig.axes[0]
ax.set_xlabel('Mean', fontsize=20,fontweight='bold',fontname=custom_font)
ax.set_ylabel('Difference', fontsize=20,fontweight='bold',fontname=custom_font)
ax.set_title('Step Length', fontsize=20,fontweight='bold',fontname=custom_font)

plt.xticks(fontsize=15, fontname=custom_font)  # 设置X轴刻度字体大小、加粗和字体
plt.yticks(fontsize=15, fontname=custom_font)  # 设置Y轴刻度字体大小、加粗和字体


plt.subplot(1,2,2)
ax2 = pg.plot_blandaltman(data3, data4)
# 设置字体属性
font_properties = {'fontsize': 15,'fontname': custom_font}

# 修改图表中的文本元素
for text in ax2.texts:
    text.set_fontsize(font_properties['fontsize'])
    text.set_fontname(font_properties['fontname'])

#ax = fig.axes[0]
ax2.set_xlabel('Mean', fontsize=20,fontweight='bold',fontname=custom_font)
ax2.set_ylabel('Difference', fontsize=20,fontweight='bold',fontname=custom_font)
ax2.set_title('Stride Speed', fontsize=20,fontweight='bold',fontname=custom_font)
plt.xticks(fontsize=15, fontname=custom_font)  # 设置X轴刻度字体大小、加粗和字体
plt.yticks(fontsize=15, fontname=custom_font)  # 设置Y轴刻度字体大小、加粗和字体
# 显示图形
plt.show()
```
